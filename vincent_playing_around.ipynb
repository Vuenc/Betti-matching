```python
import cv2
import numpy as np
import BettiMatching

a = cv2.imread("own-examples/img7-noise.png").mean(axis=2)
b = cv2.imread("own-examples/img10-noise.png").mean(axis=2)

# I'm  also adding some extra random noise here
a = np.minimum(255, np.maximum(0, a + 10 * np.random.randn(*a.shape))).astype(int)
b = np.minimum(255, np.maximum(0, b + 10 * np.random.randn(*b.shape))).astype(int)

# inverting the images: I think we are looking for darker cycles, not brighter cycles
a, b = 255 - a, 255 - b

bM = BettiMatching.BettiMatching(a, b)

bM.plot_representative_cycles(matches=300, plot_comparison=True, dim=1)
```
